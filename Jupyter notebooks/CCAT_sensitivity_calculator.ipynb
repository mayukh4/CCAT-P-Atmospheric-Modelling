```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### modified ACT files 

#### structure ACT_annual_(i)_deg_(s)_h20_factor 

##### i : from 15 to 75 degree
##### s : h20_trop_factor from 1 to 40

##### from : ACT_annual_1.15.out ----> ACT_annual_1.75.out
##### to   : ACT_annual_40_15.out ----> ACT_annual_40.75.out

##### total : 2440 files

##### output : f (ghz), tx, Trj

```python
za = np.linspace(15,75,61)
```

```python
# for h20_factor 1 - ACT based files

tx_CCAT_ACT_annual_1 =[]
trj_CCAT_ACT_annual_1 =[]

for i in np.asarray(np.linspace(15,75,61),dtype = 'int'):
    CCAT_annual = np.loadtxt("VariablePWV/ACT_annual_1."+ str(i) + ".out")
    #f_CCAT_annual = CCAT_annual[:,0]
    tx_CCAT_annual = np.mean(CCAT_annual[:,1])
    trj_CCAT_annual = np.mean(CCAT_annual[:,2])
    
    #f_CCAT_annual_1.append(f_CCAT_annual)
    tx_CCAT_ACT_annual_1.append(tx_CCAT_annual)
    trj_CCAT_ACT_annual_1.append(trj_CCAT_annual)
```

### output f GHz  tau  tx  Trj K  Tb K

```python
# for h20_factor 1 - ACT based files

tx_CCAT_ALMA_annual_1 =[]
trj_CCAT_ALMA_annual_1 =[]

for i in np.asarray(np.linspace(15,75,61),dtype = 'int'):
    CCAT_ALMA_annual = np.loadtxt("CCAT_repo_style/am_CCAT_annual_"+ str(i) + "_1.out")
    #f_CCAT_annual = CCAT_annual[:,0]
    tx_CCAT_ALMA_annual = np.mean(CCAT_ALMA_annual[:,2])
    trj_CCAT_ALMA_annual = np.mean(CCAT_ALMA_annual[:,3])
    
    #f_CCAT_annual_1.append(f_CCAT_annual)
    tx_CCAT_ALMA_annual_1.append(tx_CCAT_ALMA_annual)
    trj_CCAT_ALMA_annual_1.append(trj_CCAT_ALMA_annual)
```

```python
fig, axs = plt.subplots(1,2,sharex=True,figsize=(18,6))
fig.suptitle('ACT annual files (as on the repo) [h2o_trop_factor = 1]',size=13)

axs[0].plot(za,tx_CCAT_ACT_annual_1,'g')
axs[0].set_title("Zenith angle vs Atmospheric transmission")
axs[0].set_ylabel("Atmospheric Transmission")


axs[1].plot(za,trj_CCAT_ACT_annual_1)
axs[1].set_title("Zenith angle vs Rayleigh Jeans temperature")
axs[1].set_ylabel("Rayleigh Jeans temperature(K)")

fig.text(0.5, 0.01, 'Zenith Angle (degrees)', ha='center',size=13)

plt.show()
```

```python
fig, axs = plt.subplots(1,2,sharex=True,figsize=(18,6))
fig.suptitle('ALMA annual files (my modifications) [h2o_trop_factor = 1]',size=13)

axs[0].plot(za,tx_CCAT_ALMA_annual_1,'g')
axs[0].set_title("Zenith angle vs Atmospheric transmission")
axs[0].set_ylabel("Atmospheric Transmission")


axs[1].plot(za,trj_CCAT_ALMA_annual_1)
axs[1].set_title("Zenith angle vs Rayleigh Jeans temperature")
axs[1].set_ylabel("Rayleigh Jeans temperature(K)")

fig.text(0.5, 0.01, 'Zenith Angle (degrees)', ha='center',size=13)

plt.show()
```
